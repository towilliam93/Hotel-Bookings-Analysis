### Correlations using Seaborn

```python
# Creating a subplot removing some date information
f, ax = plt.subplots(figsize=(18,18))
df_subset = df_bookings2[['is_canceled','lead_time','arrival_date_week_number','stays_in_weekend_nights','stays_in_week_nights',
                         'adults','children','babies','is_repeated_guest','previous_cancellations','previous_bookings_not_canceled',
                         'booking_changes','days_in_waiting_list','average_daily_rate','required_car_parking_spaces','total_of_special_requests']]


# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.
corr = sns.heatmap(df_subset.corr(), annot = True, ax = ax)
# The `annot` argument allows the plot to place the correlation coefficients onto the heatmap.
```

```python
corrplot = corr.get_figure()
corrplot.savefig("seaborn_correlation.png", bbox_inches='tight')
```

- There are several notable correlations on the correlation matrix.
- Lead Time and Stays in Weekend Nights / Stays in Week Nights, with 0.24 and 0.31 respectively.
    - The larger the lead time, the higher the number of days stayed at the hotel. Possibly because people are booking longer stays further ahead in advance for availability.
- Stays in Weekend Nights and Stays in Week Nights, with 0.55.
    - The higher the amount of week nights people stay, the higher the amount of weekend nights people stay. 
- Average Daily Rate and Adults/Children, with 0.24 and 0.33 respectively.
    - More people means higher costs.
- Is Repeated Guest and Previous Cancellations/Previous Bookings Not Cancelled, with 0.21 and 0.45 respectively.
    - A repeat guest is more likely to have more bookings.
- Previous Cancellations and Previous Bookings Not Canceled, at 0.4.
    - A guest that has cancelled before is more likely to have bookings that have not been cancelled as well.

### Categorical Plots

```python
# Using a histogram to visualize distribution of arrival_date_week_number
# adding "kde" allows for distribution line to show.
sns.histplot(df_bookings2['arrival_date_week_number'], bins = 53, kde = True)
```

```python
# Creating new columns for seasons. Because Portugal is the location of the hotels, 
# we can make a rough estimate by going with Northern Hemisphere Meteorological seasons
# Spring is from March 1st to May 31st
# Summer is from June 1st to August 31st
# Autumn is from September 1st to November 31st
# Winter is from December 1st to February 28th
```

```python
df_bookings2.loc[(df_bookings2['arrival_date_week_number'] >= 9) & (df_bookings2['arrival_date_week_number'] < 22), 'Season'] = 'Spring'
```

```python
df_bookings2.loc[(df_bookings2['arrival_date_week_number'] >= 22) & (df_bookings2['arrival_date_week_number'] < 35), 'Season'] = 'Summer'
```

```python
df_bookings2.loc[(df_bookings2['arrival_date_week_number'] >= 35) & (df_bookings2['arrival_date_week_number'] < 48), 'Season'] = 'Fall'
```

```python
df_bookings2.loc[(df_bookings2['arrival_date_week_number'] >= 48) & (df_bookings2['arrival_date_week_number'] < 54), 'Season'] = 'Winter'
```

```python
df_bookings2.loc[(df_bookings2['arrival_date_week_number'] >= 0) & (df_bookings2['arrival_date_week_number'] < 9), 'Season'] = 'Winter'
```

```python
df_bookings2['Season'].value_counts(dropna= False)
```

```python
sns.set(style="ticks")
cat_plot = sns.catplot(x="arrival_date_week_number", y="stays_in_weekend_nights", hue="Season", data=df_bookings2, height=3, aspect=4)
```

- Here I created a categorical plot breaking down the length of weekend stays by seasons, based on the week that guests arrive.
- With this chart, we can see that there are more guests staying for more weekends during the winter season, especially during the first few weeks of each year. 

```python
cat_plot2 = sns.catplot(x="arrival_date_week_number", y="adults", hue="Season", data=df_bookings2, height=3, aspect=4)
```

- I also noticed in the pairplot that there was a spike in adult visitors during the fall. Perhaps there is a festival of some sort that goes on in the areas around the hotels during this time.
